```python
import tensorflow as tf
import tensorflow.keras.layers as layers 
tf.enable_eager_execution()
print(tf.__version__)
```

```python
class Mnist(tf.keras.Sequential):
    def __init__(self):
        super(Mnist, self).__init__()
        self.conv1 = layers.Conv2D(16,[3,3], activation='relu')
        self.conv2 = layers.Conv2D(16,[3,3], activation='relu')
        self.conv3 = layers.Conv2D(16,[3,3], activation='relu')
        self.flat = layers.Flatten()
        self.dense = layers.Dense(10, activation='softmax')

    def __call__(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.flat(x)
        x = self.dense(x)
        return x
```

```python
model = Mnist()
print(model)
random_state = 42
```
